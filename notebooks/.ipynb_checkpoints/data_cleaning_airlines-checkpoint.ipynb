```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
AIRLINES_FILE = "../data/airlines.csv"
AIRLINES_FILE_CLEANED_CSV = "../data/cleaned/cleaned_earth_airlines.csv"
AIRLINES_FILE_CLEANED_JSON = "../data/cleaned/cleaned_earth_airlines.json"
```

# Read Airline Data

```python
airlines_df = pd.read_csv(AIRLINES_FILE, 
#                        index_col = 0, 
                         header=0)
```

# Airline Data Cleaning

```python
airlines_df = airlines_df[airlines_df.Active == 'Y']
```

```python
airlines_df = airlines_df[airlines_df.IATA != '\\N']
```

```python
airlines_df.drop(["Airline_ID","Alias","ICAO","Callsign","Active"], inplace = True, axis =1)
```

```python
list(airlines_df.columns.values)
```

```python
airlines_df.dropna(inplace = True)
```

```python
airlines_df.reset_index(drop= True, inplace=True)
airlines_df
```

# Create JSON Data

```python
airlines_json = {}

airlines_count = len(airlines_df)

for airline_index in range(airlines_count):
    airline_data = airlines_df.iloc[airline_index]
    airlines_json[airline_data["IATA"]] = dict(airline_data)
```

# Airline Visulization

```python
no_of_countries = len(airlines_df["Country"].value_counts())
print("number of countries:", no_of_countries)
```

```python
plt.close("all")
plt.figure()
name_plt = airlines_df["Country"].value_counts().plot(kind='bar',figsize=(30,5))
name_plt.set_xlabel("Country")
name_plt.set_ylabel("No. of Airlines")
plt.show()
```

# Save Airline Data

```python
airlines_df.to_csv(AIRLINES_FILE_CLEANED_CSV)
```

```python
with open(AIRLINES_FILE_CLEANED_JSON, 'w') as fp:
    json.dump(airlines_json, fp)
```
