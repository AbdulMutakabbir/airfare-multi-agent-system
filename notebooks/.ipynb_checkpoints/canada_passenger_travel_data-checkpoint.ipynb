```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

# Reading dataset

```python
CANADA_PASSANGER_STAT_FILE_LOC = "../data/canada_filght_passenger_inbound_outbound_data.csv"
```

```python
# Read CSV files into DataFrame df

df_canada_passanger_stats = pd.read_csv(CANADA_PASSANGER_STAT_FILE_LOC, 
#                                         index_col = 0, 
                                        header=0)
```

```python
df_canada_passanger_stats.describe()
```

# Cleaning Data
removing unnecessarry columns

```python
df_canada_passanger_stats.drop(["DGUID", "UOM","UOM_ID","SCALAR_FACTOR","SCALAR_ID","VECTOR","STATUS","SYMBOL","TERMINATED","DECIMALS"], inplace = True, axis =1)
```

```python
df_canada_passanger_stats = df_canada_passanger_stats[df_canada_passanger_stats["GEO"].str.contains(',') == True]
```

```python
df_canada_passanger_stats['GEO'] = df_canada_passanger_stats['GEO'].transform(lambda  row: row.split(",",1)[0])
```

```python
df_clean_group_by = df_canada_passanger_stats.groupby(["REF_DATE","Air passenger traffic","GEO"]).sum()
```

```python
df_clean_group_by.sort_values(by=['REF_DATE','Air passenger traffic','VALUE'], inplace=True, ascending=False)
```

```python
df_clean_group_by
```

# Extracting yearly travel data

```python
df_yearly = {}

START_YR = 2013
END_YR = 2020

for year in range(START_YR, END_YR+1):
    passanger_data = {
        'arrival': None,
        'departure': None,
        'domestic': None,
        'complete': None,
    }
    passanger_data['arrival'] = df_clean_group_by.loc[(year,'Passengers deplaned')][["VALUE"]]
    passanger_data['departure'] = df_clean_group_by.loc[(year,'Passengers enplaned')][["VALUE"]]
    passanger_data['domestic'] = df_clean_group_by.loc[(year,'Domestic sector')][["VALUE"]]/2
    passanger_data['complete'] = passanger_data['arrival'].join(passanger_data['departure'], 
                                                                how="inner", 
                                                                lsuffix="_ARRIVAL", 
                                                                rsuffix="_DEPARTURE")
    passanger_data['complete'] = passanger_data['complete'].join(passanger_data['domestic'], 
                                                                 how="inner", 
                                                                 lsuffix="", 
                                                                 rsuffix="")
    passanger_data['complete'].rename(columns = {'VALUE':'VALUE_DOMESTIC'}, inplace = True)
    df_yearly[year] = passanger_data

# df_yearly[START_YR]['complete']
```

```python
df_year_comp = df_yearly[START_YR]['complete'].join(df_yearly[START_YR + 1]['complete'], 
                                             how="inner", 
                                             lsuffix="_"+str(START_YR), 
                                             rsuffix="_"+str(START_YR + 1))

for year in range(START_YR + 1, END_YR + 1):
    df_year_comp = df_year_comp.join(df_yearly[year]['complete'],
                                      how="inner",
                                      lsuffix="",
                                      rsuffix="_"+str(year))
```

```python
df_year_comp[:10].plot(kind='bar',figsize=(50,20))
```

```python
arrivals_col = []
for year in range(START_YR, END_YR+1):
    col_name = "VALUE_ARRIVAL_" + str(year)
    arrivals_col.append(col_name)

df_year_comp[:10][arrivals_col].plot(kind='bar',figsize=(50,20))
```
