```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
AIRPORTS_FILE = "../data/airports.csv"
AIRPORTS_FILE_CLEANED_CSV = "../data/cleaned/cleaned_earth_airports.csv"
AIRPORTS_FILE_CLEANED_JSON = "../data/cleaned/cleaned_earth_airports.json"
```

# Read Airport Data

```python
airports_df = pd.read_csv(AIRPORTS_FILE, 
#                        index_col = 0, 
                         header=0)
```

# Airport Data Cleaning

```python
airports_df.drop(['Airport_ID','ICAO','Timezone','DST','Tz','Type','Source'], inplace = True, axis =1)
```

```python
list(airports_df.columns.values)
```

```python
airports_df = airports_df[airports_df.IATA != '\\N']
```

```python
airports_df.dropna(inplace = True)
```

```python
airports_df.reset_index(drop= True, inplace=True)
airports_df
```

# Create JSON Data

```python
airports_json = {}

airports_count = len(airports_df)

for airport_index in range(airports_count):
    airport_data = airports_df.iloc[airport_index]
    airports_json[airport_data["IATA"]] = {
        "name" : airport_data["Name"],
        "city" : airport_data["City"],
        "country" : airport_data["Country"],
        "iata" : airport_data["IATA"],
        "lat" : float(airport_data["Latitude"]),
        "long" : float(airport_data["Longitude"]),
        "altitude" : float(airport_data["Altitude"])
    }
```

# Ariport Visulization

```python
no_of_countries = len(airports_df["Country"].value_counts())
print("number of countries:", no_of_countries)
```

```python
plt.close("all")
plt.figure()
name_plt = airports_df["Country"].value_counts().plot(kind='bar',figsize=(30,5))
name_plt.set_xlabel("Country")
name_plt.set_ylabel("No. of Airports")
plt.show()
```

# Save Airport Data

```python
airports_df.to_csv(AIRPORTS_FILE_CLEANED_CSV)
```

```python
with open(AIRPORTS_FILE_CLEANED_JSON, 'w') as fp:
    json.dump(airports_json, fp)
```
